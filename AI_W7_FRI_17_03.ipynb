```python
# Load library
import pandas as pd 
import numpy as np

# Create URL 
url = 'mtcars.csv'

# Load data
data = pd.read_csv(url)

data.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
# mpq (số dặm đi trên 1 calong)
# disp (thể tích buồng đốt)
# hp (sức ngựa)
# wt
### Giá trị tuyệt đốt của cái nào lớn nhất thì cái đó phụ thuộc lớn nhất 
## Gía trị âm là tỉ lệ nghịch, ngược lại là tỉ lệ thuận
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
# Biểu thị bảng dưới màu sắc
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9),color = 'red')
### figsize = (9,9) -> 9*9 inch
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg']) #fit là hàm train cho data
print('Giá trị Intercep là: ',model.intercept_)
print('Giá trị Coeffient là: ',model.coef_)
```

```python
model.score(X = pd.DataFrame(data['wt']), y = data['mpg'])
# score: Độ chính xác 
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'green')
y_pred = model.predict(X = pd.DataFrame(data['wt']))  #pred: Mũ
plt.plot(data['wt'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T #Tranpose: chuyển vị
model.fit(X, y = data['mpg'])
print('Hệ số Intercept:', model.intercept_)
print('Hệ số Coeffient:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['mpg']))
### y = -13.38x + 1.17(x^2) + 49.93
### Accurary = 81%
```

```python
range = np.arange(1.5, 5.5, 0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'blue', xlim = (0,7))
plt.plot(range, y_pred, color = 'red')
```
